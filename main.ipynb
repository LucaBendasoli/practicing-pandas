Selecting rows on a DataFrame

```python
df.loc[0] # Returns row as a Series
```

```python
df.loc[[0, 13, 35, 67, 88, 90]] # Returns rows as a DataFrame
```

```python
df.loc[[0,13,35,67,88,90], ['Name', 'Age']] # Returns values of the choosen indexes and columns as a DataFrame
```

```python
df.iloc[:] # Returns all rows
```

```python
df.iloc[:-1] # Returns all rows except the last one
```

```python
df.iloc[:-1, 2:9] # Returns specified rows and columns range
```

I'm going to generate a new DataFrame so we can work with numbers

```python
import numpy as np
from numpy import random

num_df = pd.DataFrame(random.randn(5,5), columns=['A', 'B', 'C', 'D', 'E']) # Create new DataFrame with random numerical values
```

```python
num_df
```

```python
num_df + num_df # Adds the values of the DataFrames
```

```python
num_df > 0 # Returns wich values are greater than 0
```

```python
num_df[num_df > 0] # Returns the values of the DataFrame that are greater than 0
```

```python
num_df[num_df['C'] > 0] # Returns rows whhere the values on the column 'C' are greater than 0
```

```python
(num_df['C'] > 0) & (num_df['D'] < 0) # Returns a boolean Series with the condition
```

```python
num_df[(num_df['C'] > 0) & (num_df['D'] < 0)] # Apllying multiple conditions to filter the DataFrame
```
