Practicing core functions of pandas basic structures with a dataset I generated with chatGPT.

Let's get our data into a DataFrame and start practicing!

```python
import pandas as pd
df = pd.read_csv('pandas_practice_dataset.csv')
df
```

```python
ser1 = df['Name'] # Returns a series
```

Series are index oriented, so we can access the value of the cell by selecting them by their index

```python
ser1[0] # Returns a value
```

```python
ser1[[0, 13, 35, 67, 88, 90]] # Returns a series with the values of the choosen indexes
```

Adding and Deleting columns from a DataFrame

```python
df['new_column'] =  df['Age'] + 10 # Creates a new column
df
```

```python
df2 = df.drop(columns=['new_column']) # Returns new DataFrame with dropped column without changing the original DataFrame
df2
```

```python
del df['new_column'] # Drops a column on the original DataFrame
df
```

Selecting rows on a DataFrame

```python
df.loc[0] # Returns row as a Series
```

```python
df.loc[[0, 13, 35, 67, 88, 90]] # Returns rows as a DataFrame
```

```python
df.loc[[0,13,35,67,88,90], ['Name', 'Age']] # Returns values of the choosen indexes and columns as a DataFrame
```

```python
df.iloc[0, 1] # Returns a value searching by its position on the DataFrame
```

```python
df.iloc[:] # Returns all rows
```

```python
df.iloc[:-1] # Returns all rows except the last one
```

```python
df.iloc[:-1, 2:9] # Returns specified rows and columns range
```
